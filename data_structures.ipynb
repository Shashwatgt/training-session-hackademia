```python
# List, Tuple, Set, Dictionaries:: Python Collections(Arrays)


# List is a collection which is ordered and changeable. Allows duplicate members.
# Tuple is a collection which is ordered and unchangeable. Allows duplicate members.
# Set is a collection which is unordered, unchangeable*, and unindexed. No duplicate members.
# Dictionary is a collection which is ordered** and changeable. No duplicate members.
```

```python
thislist = ["apple", "banana", "cherry", "kiwi"]
print(thislist[2:])
```

```python
thislist = ["apple", "banana", "cherry", "orange", "kiwi", "melon", "mango"]
print(thislist[2:5])
```

```python
thislist.append("Litchi")
```

```python
# insert into specific index
thislist.insert(1, "orange")
```

```python
# to append elements from another list to the current list, we use extend() method
tropical = ["pineapple", "mango", "papaya"]
thislist.extend(tropical)
thislist
```

```python
# remove items from list
thislist.pop(3)
```

```python
# list comprehension
fruits = ["apple", "banana", "cherry", "kiwi", "mango"]
newlist = [x for x in fruits if "a" in x]
print(newlist)

```
